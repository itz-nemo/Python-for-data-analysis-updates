```python
mean_salary = df.groupby('job_title_short')['salary_year_avg'].mean()
median_salary = df.groupby('job_title_short')['salary_year_avg'].median()

# Combine into a single DataFrame
compare = pd.DataFrame({
    'Mean': mean_salary,
    'Median': median_salary
})

# Calculate the difference (Mean - Median)
compare['Difference'] = compare['Mean'] - compare['Median']

# Sort by largest difference (optional)
compare = compare.sort_values('Difference', ascending=False)

print(compare)
```
